```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import sklearn
import pickle
```

# **Cleaning Data**

```python
df=pd.read_csv('UpdatedResumeDataSet.csv')
df['Resume'][0]
```

```python
def clean_text(text):
  clean_txt = re.sub(r'http\S+', '', text) # Remove URLs
  clean_txt = re.sub(r'@[A-Za-z0-9]+', '', clean_txt) # Remove mentions
  clean_txt = re.sub(r'#', '', clean_txt) # Remove hashtags
  clean_txt = re.sub(r'[^A-Za-z0-9\s]+', '', clean_txt) # Remove special characters and punctuation
  clean_txt = re.sub(r'RT|cc|CC|rt', '', clean_txt)
  clean_txt=re.sub(r'\s+', ' ', clean_txt) # Remove extra spaces
  return clean_txt
```

```python
df['Resume']=df['Resume'].apply(lambda x: clean_text(x))
df['Resume'][0]

#before converting into numbers
df['Category'].unique()
```

# **Converting Categories into numbers**

```python
le=sklearn.preprocessing.LabelEncoder()
df['Category']=le.fit_transform(df['Category'])

#after converting into numbers
print(df['Category'].unique())
df.head()
```

# **Vectorization**

```python
tfidf = sklearn.feature_extraction.text.TfidfVectorizer(stop_words='english')
resume_tfidf = tfidf.fit_transform(df['Resume'])
print(resume_tfidf)
```

# **Splitting**

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(resume_tfidf, df['Category'], test_size=0.2, random_state=42)
```

# **Machine Learning Model**

```python
#Model#1

clf=sklearn.neighbors.KNeighborsClassifier(n_neighbors=5)
clf.fit(x_train, y_train)
ypred=clf.predict(x_test)
print(ypred)
print(sklearn.metrics.accuracy_score(y_test, ypred))
```

# **Saving Model**

```python
pickle.dump(tfidf, open('tfidf.pkl', 'wb'))

pickle.dump(clf, open('clf.pkl', 'wb'))

pickle.dump(le, open('le.pkl', 'wb'))
```
